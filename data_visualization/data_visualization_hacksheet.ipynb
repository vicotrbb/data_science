# Visualizacao de dados

Este notebook tem como finalidade compilar os diversos tipos de formas de se apresentar e visualizar dados utilizando o python, bem como oferecer um exemplo pratico rapido de sua utilizacao.

## Autores
Victor Bona - https://www.linkedin.com/in/victorbona/

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_boston

boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['Target'] = boston_dataset.target
sns.set(rc={'figure.figsize': (11.7, 8.27)})
boston
```

# Mapa de correlacao

Mapa de correlacao ou heatmap em ingles, normalmente utilizado como o nome diz, para encontrar correlacoes entre um conjunto de variaveis.

```
matrix_corrlation = boston.corr().round(2)
sns.heatmap(data=matrix_corrlation, annot=True)
plt.show()
```

# Mapa de espalhamento

Mapa de espalhamento ou scatter plot, e um grafico de pontos em um plano cartesiano, normalmente utilizado para entender a relacao entra duas variaveis.

```
plt.clf()
plt.figure(figsize=(20,5))
features = ['LSTAT', 'RM']
target = boston['Target']

for i, col in enumerate(features):
	plt.subplot(1, len(features), i+1)
	x = boston[col]
	y = target
	plt.scatter(x, y)
	plt.title(col)
	plt.xlabel(col)
	plt.ylabel('Target')
plt.show()
```

# Grafico de linhas 

Graficos de linhas sao utilizados normalmente para acompanhar a evolucao de dados em uma serie.

```
plt.plot(boston['DIS'])
plt.xlabel('Idx registro')
plt.ylabel('DIS')
plt.title('idx X dis')
plt.show()
```

# Grafico de pizza

Grafico de pizza ou pie chart, e um grafico normalmente utilizado para visualizar a distribuicao de classes de uma variaveil.

```
# Exemplo nao muito bonito, mas funcional para entender.
plt.pie(boston['CHAS'], labels=boston['CHAS'])
plt.show()
```

# Grafico de distribuicao

Grafico de distribuicao ou distplot, normalmente utilizado para visualizacao da distribuicao de classes de uma variavel.

```
sns.distplot(boston['Target'], bins=30)
plt.show()
```

# Graficos de barras

Grafico de barras e normalmente utilizado para comparar dados ou apresentar a evolucao de dados em uma serie.


```
x = np.arange(5) # "classes"
y = [10, 20, 40, 80, 160] # "ranges"
plt.figure(figsize=(15,5))

plt.subplot(1, 2, 1) # Usar len(features) no segundo parametro para vizualizacoes quando houverem mais classes a serem apresentadas
plt.bar(x, y)

plt.subplot(1, 2, 2)
plt.barh(x, y)

plt.show()
```

# Linha do tempo

Usado para apresentar dados em uma linha temporal linear

https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/timeline.html#sphx-glr-gallery-lines-bars-and-markers-timeline-py

```
dates = [2010, 2012, 2014, 2016, 2018] # Imaginary data
levels = boston['Target'].sample(5)

fig, ax = plt.subplots(figsize=(8.8, 4), constrained_layout=True)
ax.set(title="Boston bars example")

plt.stem(dates, levels, linefmt="C3-", basefmt="k-", use_line_collection=True)
plt.show()
```

# Tipos de linhas disponiveis no matplotlib

Copiado de: https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-py

```
linestyle_str = [
     ('solid', 'solid'),      # Same as (0, ()) or '-'
     ('dotted', 'dotted'),    # Same as (0, (1, 1)) or '.'
     ('dashed', 'dashed'),    # Same as '--'
     ('dashdot', 'dashdot')]  # Same as '-.'

linestyle_tuple = [
     ('loosely dotted',        (0, (1, 10))),
     ('dotted',                (0, (1, 1))),
     ('densely dotted',        (0, (1, 1))),

     ('loosely dashed',        (0, (5, 10))),
     ('dashed',                (0, (5, 5))),
     ('densely dashed',        (0, (5, 1))),

     ('loosely dashdotted',    (0, (3, 10, 1, 10))),
     ('dashdotted',            (0, (3, 5, 1, 5))),
     ('densely dashdotted',    (0, (3, 1, 1, 1))),

     ('dashdotdotted',         (0, (3, 5, 1, 5, 1, 5))),
     ('loosely dashdotdotted', (0, (3, 10, 1, 10, 1, 10))),
     ('densely dashdotdotted', (0, (3, 1, 1, 1, 1, 1)))]


def plot_linestyles(ax, linestyles):
    X, Y = np.linspace(0, 100, 10), np.zeros(10)
    yticklabels = []

    for i, (name, linestyle) in enumerate(linestyles):
        ax.plot(X, Y+i, linestyle=linestyle, linewidth=1.5, color='black')
        yticklabels.append(name)

    ax.set(xticks=[], ylim=(-0.5, len(linestyles)-0.5),
           yticks=np.arange(len(linestyles)), yticklabels=yticklabels)

    for i, (name, linestyle) in enumerate(linestyles):
        ax.annotate(repr(linestyle),
                    xy=(0.0, i), xycoords=ax.get_yaxis_transform(),
                    xytext=(-6, -12), textcoords='offset points',
                    color="blue", fontsize=8, ha="right", family="monospace")


fig, (ax0, ax1) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [1, 3]},
                               figsize=(10, 8))

plot_linestyles(ax0, linestyle_str[::-1])
plot_linestyles(ax1, linestyle_tuple[::-1])

plt.tight_layout()
plt.show()
```
