# PIPELINES

Pipelines sao formas de simplificar a execucao de uma sequencia de passos, criando uma sequencia de execucao a ser realizada antes de passarmos os dados por um modelo de regressao linear por exemplo.

Criamos um pipeline que pode efetuar passos de normalizacao e transformacao de nossos dados antes treinarmos nosso modelo de regressao linear, ou outros tipos de modelos de ML

O exemplo a seguir cria um pipeline que ira:
* Normalizar os dados;
* Passar por uma transformacao polinomial
* Executar nosso modelo de regressao linear

```
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline


df = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv')
Input=[('scale', StandardScaler()), ('Polynomial', PolynomialFeatures(degree=2)), ('mode', LinearRegression())]

x = df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
y = df['price']

pipe = Pipeline(Input)
pipe.fit(x, y)

predict = pipe.predict(x)
predict                        
```
