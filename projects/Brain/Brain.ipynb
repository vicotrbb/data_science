## Tipos de ondas cerebrais

As ondas cerebrais são produzidas pelos pulsos eletricos emitidos pelo nosso cerebro e podem ser medidas em Hz. Existem 4 tipos de ondas cerebrais basicas conhecidas, sendo elas: Beta, Alfa, Teta, Delta.

Essas ondas cerebrais são produzidas e melhor visualizadas dependendo de algum estado que o objeto de pesquisa se encontre, tais estados serão descritos a frente, mas utilizando estas leituras é possivel indentificar varios atributos e situações em que o individuo se encontra.

### Beta

A onda cerebral representa os estados relacionados as altas taxa de atenção, concentração e cognição. Podemos observar fortemente esta classe de onda durante o estado de vigília, fora a grande incidência durante crises de pânico e ansiedade(Provavelmente decorrente do alta atividade cerebral proporcionada pelo estado mental). Estas ondas são imprescindíveis para os procedimentos que exigem grande esforço mental.

* Frequência: 13 - 30 Hz.

### Alfa

Presente em estados de relaxamento, observação, meditação e visualização. As ondas Alfa são responsaveis por um estado de profundo relaxamento, meditação ou de oração. Nesse nivel que a inteligencia, memória, criatividade, inspiração, percepção sensorial e intuição se encontram.

* Frequência: 7 - 13 Hz.

### Teta

As ondas Teta estão pretens em estados de meditação profunda, intuição/criatividade e memória. Normalmente as ondas Teta são geradas no processo anterior ao de entrar no sono profundo e é gerada pela nossa mente inconsciente. É em Teta que podemos realizar uma conexão profunda com nossa espiritualidade. 

* Frequência: 4 - 7 Hz.

### Delta

As ondas Delta estão normalmente relacionadas aos estados mentais de consciencia expandida e recuperação, é a frequencia mais baixa que pode ser lida no cerebro humano e normalmente está ligada ao sono profundo. Algumas frequencias Delta podem liberar um hormônio do crescimento humano, o HGH, que tambem é bastante benefico para a restruturação celular.

* Frequência: 0 - 4 Hz.

## Interfaces cerebro - computador

### Eletroencefalograma

O Eletroencefalograma(EEG) é um equipamento que consegue analistar e ler a atividade eletrica do cerebro de forma seguro, indolor e não invasivo. Conhecido também como eletroencefalografia ou EEG.

O método de Eletroencefalografia amplifica os sinais eletricos do cerebro, possibilitando o registro desta atividade e a futura analista computadorizada. Normalmente estas informações são lidas e interpretadas pelos médicos especializados em formato de grafico de linhas, o suporte computadorizado tambem é utilizado para a melhor analise dessas informações.

Normalmente o eletroencefalograma é utilizado para identificar anomalidades nas ondas cerebrais do cerebro, por meio da analista da frequencia destas ondas. Acontece que mesmo em pessoas completamente saudaveis, o EEG pode sofrer variações dependendo do estado(idade, nivel de cansaço, fome e etc) do corpo ou de estimulos externos. 

### Como funciona um eletroencefalograma

O aparelho utilizado no eletroencefalograma é formado por um conjunto de eletrodos conectados a um computador. Durante o procedimento estes eletrodos são posicionados em diferentes posições da cabeça afim de ler e identificar a atividade eletrica na area que o eletrodo está posicionada. Vale ressaltar que este processo não é invasivo, apesar de existirem métodos que utilizam uma abordagem mais invasiva, aplicando os eletrodos diretamente no cerebro, atraves de um procedimento cirurgico.

As oscilações criadas pelos impulsos eletricos criam frequencias especificas, normalmente, neuronios que oscilam na mesma frequencia formam grupos e são "lidos" juntamente. Afim de estimular diferentes respostas destes neuronios durante o EEG, são provocados estímulos luminosos em determinadas frequências, para receber respostas dos neurônios que vibram nestas frequencias especificas. Por fim as leituras são processadas por um computador para que sejam interpretadas pelo especialista.

### Tipos de eletroencefalogramas

Como mencionado a cima o eletroencefalograma é influenciado pelo estado da pessoa que está passando pelo procedimento, este fato criou margem para o desenvolvimento de diferentes tipos de aparelhos para situações especificas. Atualmente existem 3 tipos principais de eletroencefalograma: EEG em vigília, em sono e com mapeamento cerebral.

#### Eletroencefalograma em vigília

A mais comum das versões, o Eletroencefalograma em vigília é utilizado com o paciente acordado e é utilizado para analisar a atividade espontanea do cerebro. Dependendo da situação, pode ser solicitado ao paciente que ele execute certas ações, visando o mapeamento do cerebro durante o processo, para melhor analise.

#### Eletroencefalograma em sono

Como o nome sugere, esta modalidade é utilizada com o paciente dormindo e visa analisar e detectar disturbios do sono, normalmente o paciente passa a noite inteira com o aparelho, sendo as vezes, necessario que seja sedado para garantir a continuidade do sono.

#### Eletroencefalograma com mapeamento cerebral

O grande diferencial desse modelo é a capacidade de quantificar os dados extraidos em imagens computadorizadas, capacitanto o médico a visualizar melhor quais areas estão sendo afetadas durante algum procedimento/experimento/interação/estimulo.

![mapeamento cerebral](https://telemedicinamorsch.com.br/wp-content/uploads/2017/03/Mapeamento-cerebral-700x394.jpg)

### Oque pode ser detectado com o eletroencefalograma?

Clinicamente o eletroencefalograma pode detectar uma sequencia de doenças, tais como: desordens neurológicas, doenças degenerativas do sistema nervoso, tumores e demências. Normalmente um exame de eletroencefalograma é solicitado após algum acidente que tenha atingido a cabeça, contudo, outro uso clinico extremamente importante para o eletroencefalograma é o acompanhamento de epilepsia, ou de outras causas de convulsão.

Ondas cerebrais tambem podem ser observadas atraves de um exame por eletroencefalograma.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

readings = pd.read_csv('drive/My Drive/Brain/Electroencephalogram-dataset/readings.csv', sep=';', header=0)
target = pd.read_csv('drive/My Drive/Brain/Electroencephalogram-dataset/spindles.csv', sep=';', header=0)

readings = pd.DataFrame(readings)
target = pd.DataFrame(target)

readings['Time'] = np.arange(0,248440*0.005,0.005)
readings = readings[['Time',"EOG Left", "EEG C3-A1", "EEG O1-A1", "EEG C4-A1", "EEG O2-A1"]]

target['Time'] = np.arange(0,2598*0.005,0.005)
target = target[['Time',"EOG Left[uV]", "EEG C3-A1[uV]", "EEG O1-A1[uV]", "EEG C4-A1[uV]", "EEG O2-A1[uV]"]]
```

### Leituras do EEG

```
fig, axs = plt.subplots(5, sharex=True, sharey=True)
fig.set_size_inches(18, 24)
labels = ["EOG Left", "EEG C3-A1", "EEG O1-A1", 'EEG C4-A1', 'EEG O2-A1']
colors = ["r","g","b",'y',"k"]
fig.suptitle('Leituras do EEG', fontsize=20)
# ---- loop over axes ----
for i,ax in enumerate(axs):
  axs[i].plot(readings['Time'], readings[labels[i]],color=colors[i],label=labels[i])
  axs[i].legend(loc="upper right")

plt.xlabel('Time (s)', fontsize=20)
plt.show()
```

```
fig, ax = plt.subplots(1)
fig.set_size_inches(18, 24)
labels = ["EOG Left", "EEG C3-A1", "EEG O1-A1", 'EEG C4-A1', 'EEG O2-A1']
colors = ["r","g","c",'y',"k"]
fig.suptitle('Leituras do EEG estacadas', fontsize=20)
# ---- loop over axes ----
for i,lbl in enumerate(labels):
  ax.plot(readings['Time'], readings[lbl],color=colors[i],label=lbl)

ax.legend(loc="upper right")
plt.xlabel('Time (s)', fontsize = 20)
plt.show()
```
